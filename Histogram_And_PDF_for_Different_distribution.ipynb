# Check whether Distribution is Normal or not

```python
#### If we want to check whether feature is gaussian or normal distributed, we used---
#### Q-Q Plot
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
```

```python
def plot_data(sample):
  plt.figure(figsize=(10,6))
  plt.subplot(1,2,1)
  sns.histplot(sample)
  plt.subplot(1,2,2)
  stat.probplot(sample,dist='norm',plot = pylab)
  plt.show
```

```python
## Create a normal Distributed data
s = np.random.normal(0.5,0.2,1000)
plot_data(s)
```

```python

plot_data(df['sepal_length'])
```

```python
plot_data(df['sepal_width'])
```

```python
## check whether lognormal or not
```

```python
def plot_data_log(sample):
  plt.figure(figsize=(10,6))
  plt.subplot(1,2,1)
  sns.histplot(sample)
  plt.subplot(1,2,2)
  stat.probplot(np.log(sample),dist='norm',plot = pylab)
  plt.show
```

```python
mu, sigma = 3., 1. # mean and standard deviation
x = np.random.lognormal(mu, sigma, 1000)
plot_data_log(x)
```
